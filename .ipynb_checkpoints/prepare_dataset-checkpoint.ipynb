```python
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 13, 10

import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
df1 = pd.read_csv(r'data2\weather\seattleWeather_1948-2017.csv')
df1 = df1[['DATE']+['TMIN']+['TMAX']]
df1['DATE'] = pd.to_datetime(df1['DATE'])
df1.set_index('DATE', inplace=True)
df1
```

```python
df1 = df1[df1.columns].resample('MS').mean()
df1 = df1.iloc[396:793]
```

```python
df1
```

```python
df2 = pd.read_csv(r'data/final_data2.csv')
df2.set_index('DATE', inplace=True)
df2 = df2.drop('MIN_TEMPR', axis=1)
df2
```

```python
df1.index = df2.index
df = pd.concat([df1, df2], axis = 1)
```

```python
df
```

```python
data = []
for i in df.columns:
    data.append(go.Scatter(
            x = df.index, y = df[i], name=i
            ))
    
data = data
layout = {'title': 'Data1'}

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df.to_csv(r'data\final_data_predict.csv')
```
