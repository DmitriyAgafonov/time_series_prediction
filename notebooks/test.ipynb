```python
def moving_avarage(df, n):
    df_av = df.rolling(window=n).mean()#.dropna()
    #df_av = df_av.fillna(df_av.median())
    
    return df_av
```

```python
m_a_df = moving_avarage(df, 10)
m_a_df
```

```python
def exp_smoothing(df, alpha):
    df_ex = df.ewm(alpha=alpha, adjust=True).mean().dropna()

    return df_ex
```

```python
ex_s_df = exp_smoothing(df, 0.5)
ex_s_df
```

```python
def Holt_smoothing(frame, alpha, beta):
    start_frame = frame[['DATE']+['MIN_TEMPR']+['ELECTRIC_PROD']+['BEER_PROD']]
    #start_frame = frame[['DATE']+['PITCH']+['YAW']+['ROLL']]
    result_frame = frame[['DATE']]#.set_index('TIME', inplace=True)
    
    for i in start_frame.columns[1:]:
        fit_model = Holt(start_frame[i]).fit(smoothing_level= alpha, smoothing_slope = beta)
        fit_series = fit_model.fittedvalues
        result_frame[i] = fit_series
       
        
    #result_frame.set_index('TIME', inplace=True)
    
    return result_frame
```

```python
h_s_df = Holt_smoothing(df, 0.3, 0.3)
h_s_df
```

```python
feature_drop = ['MIN_TEMPR', 'ELECTRIC_PROD', 'BEER_PROD']
smooth_drop = ['expsmooth1', 'expsmooth2', 'expsmooth3']

dfs = [m_a_df, ex_s_df, h_s_df]
dict_dfs = dict(zip(smooth_drop, dfs))
```

```python
def plt(feature_drop, smooth_drop):
    
    data =[]
    
    for i in feature_drop[:]:
        data.append(go.Scatter(
                x = df_time, y = df[i], name=i
                ))


    for j in smooth_drop[:]:
        for i in feature_drop[:]:
            data.append(go.Scatter(
                x = df_time, y = dict_dfs[j][i], name=i+j
                ))
    
    
    layout = {'title': 'Graph'}

    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
    
```

```python
plt(feature_drop, smooth_drop)
```
