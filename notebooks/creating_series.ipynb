```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.api import Holt

import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
data1 = pd.read_csv('data2/daily-minimum-temperatures-in-me.csv')
data1 = data1.iloc[:397, :]
```

```python
data1
```

```python

trace0 = go.Scatter(
            x = data1['Date'], y = data1['Daily minimum temperatures'], name='Daily minimum temperatures'
            )
    
data = [trace0 ]
layout = {'title': 'Data1'}
    
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
data2 = pd.read_csv('data2/Electric_Production.csv')
data2
```

```python
trace0 = go.Scatter(
            x = data2['DATE'], y = data2['IPG2211A2N'], name='IPG2211A2N'
            )
    
data = [trace0 ]
layout = {'title': 'Data2'}
    
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
data3 = pd.read_csv('data2/monthly-beer-production-in-austr.csv')
data3 = data3.iloc[:397, :]
data3
```

```python
trace0 = go.Scatter(
            x = data3['Month'], y = data3['Monthly beer production'], name='Monthly beer production'
            )
    
data = [trace0 ]
layout = {'title': 'Data3'}
    
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df_time = data2['DATE']
df_time = pd.to_datetime(df_time)
```

```python
final_df = pd.concat([df_time, data1['Daily minimum temperatures'], data2['IPG2211A2N'], data3['Monthly beer production']], axis=1)
```

```python
final_df
```

```python
new_cols = ['DATE', 'MIN_TEMPR', 'ELECTRIC_PROD', 'BEER_PROD']
old_cols = final_df.columns
replace_cols = dict(zip(old_cols, new_cols))
```

```python
final_df = final_df.rename(columns = replace_cols)
```
